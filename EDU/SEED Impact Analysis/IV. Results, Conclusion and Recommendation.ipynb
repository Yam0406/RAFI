# Recommendations

- Focus improvements on the following:
    - _Assessment of Student Performance_. Train and capacitate teachers to utilize varied assessment techniques and methodologies (traditional, non-traditional or authentic).
    - _Budgeting Process_. Conduct trainings and sessions with the schools on budgeting and encourage them to consult with stakeholders to identify and prioritize current needs as well as maximize existing assets in the community.
    - _Teacher Competence_. Put teachers into internal and external trainings as part of continuous education and to improve their competence.
    - _Classroom Structure_. Increase the number of classrooms that are safe, conducive, learner-centered, well-structured, and inclusive for all types of learners (including those with special needs and disabilities).
    - _Staff Assessment System_. Help the schools establish staff assessment systems wherein results are analyzed for intervention and reflected in the priorities of the school.
    - _Asset Monitoring System_. Aid schools in the creation of an asset monitoring and maintenance system of their physical resources with proper documentation and reporting to the stakeholders.
- Perform random validation of the scores given by the community assessors and site validators and review the cause of the discrepancies or variances. Also, check whose evaluation is more accurate specially those that can be verified through quantitative data and documents (e.g. drop-out rate, asset monitoring system through various reports)
- Reassess the method of evaluation on indicators that are evaluated based on observation and subjective judgment or those that are prone to bias (e.g. teacher competence can be evaluated through tests, documentation, and/or trainings attended instead of observation of sample classes in the core subject areas).
- Acquire quantitative data from the schools including but not limited to student grades, student attendance, NAT scores, etc.
